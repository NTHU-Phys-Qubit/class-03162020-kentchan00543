```python
#Exercise 1 & 2 : 2D 
#老師給的題目HW
def my_random1(start,end,number):
    x= np.zeros(number);
    y= np.zeros(number);

    for p in range(0,number):
        x1 = 0
        y1 = 0
        for i in range(0,number):
            delta_x = np.random.random(number)*(end- start)+ start
            delta_y = np.random.random(number)*(end- start)+ start
            x1 = x1 + delta_x[i]
            y1 = y1 + delta_y[i]
            #print(x1)
        x[p] = x1
        y[p] = y1
        
        
    num_bins = 50
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate')
    plt.ylabel('Numbers of Appeared')
    plt.title('Histogram of Random Walk—— x-component ')
    plt.grid(True)
    plt.show() 
    
    num_bins = 50
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('y-coordinate')
    plt.ylabel('Numbers of Appeared')
    plt.title('Histogram of Random Walk—— y-component ')
    plt.grid(True)
    plt.show() 

    plt.title('Trajectory of Random Walk in 2 Dimensions')
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.plot(x, y)
```

```python
my_random1(start=-1,end=1,number=200)
```

```python
#Exercise 3 : 2D
#老師的第三題
def my_random2(number):
    x= np.zeros(number);
    y= np.zeros(number);
    theta1 = np.zeros(number);
    for p in range(0,number):
        x1 = 0
        y1 = 0
        L = 1;
        
        for i in range(0,number):
            theta = (np.random.random(number))*2*math.pi;
            theta1[i] = math.degrees(theta[i])
            #print(theta1)
            delta_x = L *(math.cos(theta1[i]));
            delta_y =  L *(math.sin(theta1[i]));
            x1 = x1 + delta_x
            y1 = y1 + delta_y
            #print(x1)
        x[p] = x1
        y[p] = y1
        
        
    num_bins = 50
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate')
    plt.ylabel('Numbers of Appeared')
    plt.title('Histogram of Random Walk—— x-component ')
    plt.grid(True)
    plt.show() 
    
    num_bins = 50
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('y-coordinate')
    plt.ylabel('Numbers of Appeared')
    plt.title('Histogram of Random Walk—— y-component ')
    plt.grid(True)
    plt.show() 

    plt.title('Trajectory of Random Walk in 2 Dimensions')
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.plot(x, y)
```

```python
my_random2(number=100)
```
